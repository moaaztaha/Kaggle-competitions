```python
small_train_data = pd.read_csv('data/Train.csv', delimiter=',')
small_train_data = small_train_data[['SalesID', 'MachineID', 'SalePrice', 'saledate']]
```

```python
small_train_data.saledate = [date.year for date in pd.to_datetime(small_train_data.saledate)]
```

```python
small_train_data.info()
```

```python
y = small_train_data['SalePrice']
X = small_train_data.drop('SalePrice', axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, cross_val_score

reg_rfg = RandomForestRegressor(random_state=42)
param_grid = {'bootstrap': [False, True], 'n_estimators': [3, 10], 'max_features': ['auto', 'sqrt']}

grid_search = GridSearchCV(reg_rfg, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True, n_jobs=-1)
grid_search.fit(X, y)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_
```

```python
import numpy as np
negative_mse = grid_search.best_score_
rmse = np.sqrt(-negative_mse)
rmse
```

```python
from sklearn.metrics import mean_squared_log_error
def rmsle(y, y_pred):
    score = np.sqrt(mean_squared_log_error(y, y_pred))
    print(score)
    return score
```

```python
y_pred = grid_search.predict(X)
score = rmsle(y_pred, y)
```
